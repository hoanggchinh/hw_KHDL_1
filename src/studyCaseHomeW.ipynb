```python
# Chuyen cot datetime thanh kieu dt
df["datetime"] = pd.to_datetime(df["datetime"])
```

```python
df.dtypes #check lai kieu du lieu cua cot dt
```

```python
# 3. Trich xuat thong tin thoi gian add cot y, m, dow, h
df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["day_of_week"] = df["datetime"].dt.day_name()
df["hour"] = df["datetime"].dt.hour
```

```python
# 2.Loc du lieu chi lay du lieu tu 2013 - 2016
df = df[(df["datetime"].dt.year >= 2013) & (df["datetime"].dt.year <= 2016)]
```

```python
df.iloc[:, -5:].head()
```

```python
df[df.isna().any(axis=1)] # ktra 1 lan nua xem con nhieu du lieu bi thieu khong
```

```python
# 5. noi suy du lieu bi thieu
df.iloc[:, 1:-4] = df.iloc[:, 1:-4].interpolate(method='linear') # doi cac cot tru cot datetime va 4 cot y m d h
```

```python
df[df.isna().any(axis=1)]
```

```python
df.iloc[:, 1:-4] = (df.iloc[:, 1:-4] / 10).round(2)
```

```python
df.head()
```

```python
#loc cot du thua de tinh
df_onlydt = df.drop(columns=['year', 'month', 'day_of_week', 'hour'])
```

```python
df_onlydt.head()
```

```python
# 4. chuyen t trung binh theo gio thanh t trung binh theo ngay
df_dtb = df_onlydt.resample("D", on="datetime").mean()
```

```python
df_dtb.head()
```

```python
import matplotlib.pyplot as plt

# Resample để tính nhiệt độ trung bình theo tháng
df_monthly = df_dtb.resample('ME').mean()

# Biểu đồ cho San Francisco
plt.figure(figsize=(8, 5))
plt.plot(df_monthly.index, df_monthly['San Francisco'], label='San Francisco', marker='o', color='b')
plt.xlabel('Tháng')
plt.ylabel('Nhiệt độ trung bình (°C)')
plt.title('Xu hướng nhiệt độ trung bình theo tháng (San Francisco)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Biểu đồ cho Los Angeles
plt.figure(figsize=(8, 5))
plt.plot(df_monthly.index, df_monthly['Los Angeles'], label='Los Angeles', marker='o', color='g')
plt.xlabel('Tháng')
plt.ylabel('Nhiệt độ trung bình (°C)')
plt.title('Xu hướng nhiệt độ trung bình theo tháng (Los Angeles)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Biểu đồ cho Las Vegas
plt.figure(figsize=(8, 5))
plt.plot(df_monthly.index, df_monthly['Las Vegas'], label='Las Vegas', marker='o', color='r')
plt.xlabel('Tháng')
plt.ylabel('Nhiệt độ trung bình (°C)')
plt.title('Xu hướng nhiệt độ trung bình theo tháng (Las Vegas)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
