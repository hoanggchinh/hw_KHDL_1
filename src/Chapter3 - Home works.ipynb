## **📝 Bài tập 1: Phân tích dữ liệu về xe taxi tại New York**
📥 **Dữ liệu:** [New York City Taxi Trip Data (2019)](https://www.kaggle.com/datasets/ehallmar/nyc-taxi-trip-duration-2019)

### **Yêu cầu bài tập**
1. **Đọc và khám phá dữ liệu** (`.info()`, `.head()`, `.describe()`).
2. **Xử lý dữ liệu bị thiếu**: Kiểm tra giá trị NaN, điền giá trị thiếu hợp lý.
3. **Lọc dữ liệu**: Lọc các chuyến đi có **thời gian ngắn hơn 2 phút hoặc dài hơn 1 ngày** (bất thường).
4. **Chuyển đổi cột thời gian**: Định dạng lại `pickup_datetime` thành kiểu **datetime**.
5. **Tạo thêm cột mới**: Trích xuất **giờ trong ngày**, **ngày trong tuần**, **tháng** từ cột thời gian.
6. **Nhóm dữ liệu**: Tính **thời gian trung bình của chuyến đi** theo từng tháng.
7. **Vẽ biểu đồ** thể hiện số chuyến đi theo tháng.

### **Ví dụ kết quả mong đợi**
Sau khi xử lý, DataFrame đầu ra sẽ như sau:
| pickup_datetime      | duration_min | hour | day_of_week | month |
|----------------------|-------------|------|------------|-------|
| 2019-01-01 08:30:00 | 15          | 8    | Tuesday    | 1     |
| 2019-02-15 14:45:00 | 23          | 14   | Friday     | 2     |

### **Gợi ý cách làm**
- Sử dụng `.to_datetime()` để chuyển đổi thời gian.
- Dùng `.dt.hour`, `.dt.day_name()` để trích xuất thông tin thời gian.
- Lọc dữ liệu bất thường với điều kiện `(df["duration_min"] < 2) | (df["duration_min"] > 1440)`.
- Dùng `.groupby("month")["duration_min"].mean()` để tính thời gian trung bình.

## **📝 Bài tập 2: Phân tích giá vé máy bay**
📥 **Dữ liệu:** [US Domestic Airline Flights Dataset](https://www.kaggle.com/datasets/ayushpradhaan/airline-flights-data)

### **Yêu cầu bài tập**
1. **Nối dữ liệu**: Gộp bảng `flights.csv` với bảng `airline_prices.csv` theo mã chuyến bay.
2. **Xử lý dữ liệu thiếu**: Điền giá trị NaN trong cột giá vé bằng **giá trung bình của hãng hàng không đó**.
3. **Xử lý dữ liệu chuỗi**: Chuẩn hóa tên hãng hàng không thành **chữ in thường**.
4. **Nhóm dữ liệu**: Tính giá vé trung bình của mỗi hãng hàng không.
5. **Phân tích theo thời gian**: Tính số lượng chuyến bay theo từng tháng.

### **Ví dụ kết quả mong đợi**
Sau khi xử lý, bảng thống kê giá vé trung bình theo hãng hàng không sẽ như sau:
| Airline             | Avg_Price |
|---------------------|----------|
| american airlines  | 250      |
| delta airlines     | 280      |
| united airlines    | 230      |

### **Gợi ý cách làm**
- Dùng `.merge(df1, df2, on="Flight")` để kết hợp dữ liệu.
- Xử lý NaN bằng `.fillna(df.groupby("Airline")["Price"].transform("mean"))`.
- Dùng `.str.lower()` để chuẩn hóa tên hãng hàng không.
- Dùng `.groupby("Airline")["Price"].mean()` để tính trung bình giá vé.
- Sử dụng `.resample("M")` để phân tích số lượng chuyến bay theo tháng.

## **📝 Bài tập 3: Phân tích dữ liệu đơn hàng thương mại điện tử**
📥 **Dữ liệu:** [E-commerce Data](https://www.kaggle.com/datasets/carrie1/ecommerce-data)

### **Yêu cầu bài tập**
1. **Đọc dữ liệu** và chuyển đổi cột `InvoiceDate` sang dạng datetime.
2. **Xử lý dữ liệu thiếu**: Xóa các đơn hàng không có mã sản phẩm.
3. **Lọc dữ liệu**: Loại bỏ các đơn hàng **có giá trị âm hoặc số lượng âm**.
4. **Trích xuất thông tin thời gian**: Thêm cột `month`, `hour`, `day_of_week`.
5. **Tạo Pivot Table**: Xây dựng bảng tổng hợp doanh thu theo tháng và quốc gia.
6. **Nhóm dữ liệu**: Tính số đơn hàng trung bình của từng quốc gia.

### **Ví dụ kết quả mong đợi**
Pivot Table đầu ra:
| Country       | Jan  | Feb  | Mar  | ... |
|--------------|------|------|------|-----|
| United Kingdom | 5000 | 5200 | 5300 | ... |
| Germany       | 1200 | 1300 | 1250 | ... |

### **Gợi ý cách làm**
- Dùng `.dropna(subset=["StockCode"])` để loại bỏ đơn hàng thiếu mã sản phẩm.
- Lọc dữ liệu bằng điều kiện `(df["Quantity"] > 0) & (df["UnitPrice"] > 0)`.
- Dùng `.pivot_table(values="TotalPrice", index="Country", columns="month", aggfunc="sum")` để tổng hợp doanh thu.
- Dùng `.groupby("Country")["InvoiceNo"].count()` để đếm số đơn hàng theo quốc gia.

## **📝 Bài tập 4: Phân tích dữ liệu thời tiết**
📥 **Dữ liệu:** [Hourly Weather Data](https://www.kaggle.com/datasets/selfishgene/historical-hourly-weather-data)

### **Yêu cầu bài tập**
1. **Đọc dữ liệu và kiểm tra thông tin cột thời gian**.
2. **Lọc dữ liệu**: Chỉ giữ lại dữ liệu từ năm 2020 trở đi.
3. **Trích xuất thông tin thời gian**: Thêm cột `year`, `month`, `day_of_week`, `hour`.
4. **Lấy mẫu lại dữ liệu**: Chuyển từ dữ liệu theo giờ thành **trung bình nhiệt độ theo ngày**.
5. **Nội suy dữ liệu bị thiếu** để điền nhiệt độ bị khuyết.
6. **Tính toán xu hướng thời gian**: Tính nhiệt độ trung bình theo tháng và vẽ biểu đồ.

### **Ví dụ kết quả mong đợi**
Sau khi resample, dữ liệu có dạng:
| Date       | Avg_Temp |
|------------|---------|
| 2020-01-01 | 5.2     |
| 2020-01-02 | 4.8     |
| 2020-01-03 | 3.1     |

### **Gợi ý cách làm**
- Dùng `.to_datetime(df["datetime"])` để chuyển đổi cột thời gian.
- Lọc dữ liệu bằng điều kiện `df["datetime"].dt.year >= 2020`.
- Sử dụng `.resample("D", on="datetime").mean()` để lấy nhiệt độ trung bình theo ngày.
- Dùng `.interpolate()` để nội suy nhiệt độ bị thiếu.
- Vẽ biểu đồ xu hướng nhiệt độ theo tháng bằng Matplotlib.
